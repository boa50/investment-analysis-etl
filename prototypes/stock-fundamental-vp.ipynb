```python
import pandas as pd
import os
```

```python
years_load = [2022, 2023, 2024]
# files_load = ["dfp_cia_aberta_BPP_con_", "dfp_cia_aberta_BPP_ind_", "itr_cia_aberta_BPP_con_", "itr_cia_aberta_BPP_ind_"]
files_load = ["dfp_cia_aberta_BPP_ind_", "itr_cia_aberta_BPP_ind_"]

df = pd.DataFrame()

for year in years_load:
    for file in files_load:
        fname = "../data/raw/" + file + str(year) + ".csv"

        if os.path.isfile(fname):
            df = pd.concat(
                [
                    df,
                    pd.read_csv(
                        fname,
                        encoding="ISO-8859-1",
                        sep=";",
                    ),
                ]
            )
        else:
            print(fname + " not found! Skipping")
            
df_basic_info = pd.read_csv("../data/processed/stocks-basic-info.csv")
df = df[df["CD_CVM"].isin(df_basic_info["CD_CVM"].values)]

df = df[df["ORDEM_EXERC"] == "ÚLTIMO"]
# df = df[df["DS_CONTA"].isin(["Patrimônio Líquido Consolidado", "Patrimônio Líquido"])]
df = df[df["DS_CONTA"] == "Patrimônio Líquido"]
```
