```python
import pandas as pd
import datetime
```

```python
dfp_dre_2022 = pd.read_csv("../data/raw/dfp_cia_aberta_DRE_con_2022.csv", encoding = "ISO-8859-1", sep=";")
dfp_dre_2023 = pd.read_csv("../data/raw/dfp_cia_aberta_DRE_con_2023.csv", encoding = "ISO-8859-1", sep=";")

itr_dre_2022 = pd.read_csv("../data/raw/itr_cia_aberta_DRE_con_2022.csv", encoding = "ISO-8859-1", sep=";")
itr_dre_2023 = pd.read_csv("../data/raw/itr_cia_aberta_DRE_con_2023.csv", encoding = "ISO-8859-1", sep=";")
itr_dre_2024 = pd.read_csv("../data/raw/itr_cia_aberta_DRE_con_2024.csv", encoding = "ISO-8859-1", sep=";")
df = pd.concat([itr_dre_2022, dfp_dre_2022, itr_dre_2023, dfp_dre_2023, itr_dre_2024])
```

```python
df.head()
```

```python
df["DT_FIM_EXERC"] = pd.to_datetime(df["DT_FIM_EXERC"])
df["DT_INI_EXERC"] = pd.to_datetime(df["DT_INI_EXERC"])
df["EXERC_YEAR"] = df["DT_INI_EXERC"].dt.year
```

```python
df_bbas3 = df[df["CD_CVM"] == 1023]
df_bbas3 = df_bbas3[df_bbas3["ORDEM_EXERC"] == "ÚLTIMO"]
df_bbas3["DS_CONTA"].unique()
```

```python
kips = [
    "Lucro ou Prejuízo Líquido Consolidado do Período",
    ]

disp_columns = ["EXERC_YEAR", "DT_INI_EXERC", "DT_FIM_EXERC", "DS_CONTA", "VL_CONTA"]

df_bbas3[df_bbas3["DS_CONTA"].isin(kips)][disp_columns]
```

```python
td_quarter = datetime.timedelta(days=93)
td_year = datetime.timedelta(days=360)

profit_raw = df_bbas3[df_bbas3["DS_CONTA"].isin(kips)]
dt_diff = profit_raw["DT_FIM_EXERC"] - profit_raw["DT_INI_EXERC"]

profit_quarter = profit_raw[dt_diff <= td_quarter]
profit_quarter_grouped = profit_quarter.groupby("EXERC_YEAR").agg({"VL_CONTA": "sum", "DT_FIM_EXERC": "max"}).reset_index()

profit_year = profit_raw[dt_diff >= td_year]
profit_year = pd.merge(profit_year, profit_quarter_grouped, how="inner", on="EXERC_YEAR", suffixes=("_year", "_quarters"))
profit_year["VL_CONTA"] = profit_year["VL_CONTA_year"] - profit_year["VL_CONTA_quarters"]
profit_year["DT_INI_EXERC"] = profit_year["DT_FIM_EXERC_quarters"] + datetime.timedelta(days=1)
profit_year = profit_year.rename({"DT_FIM_EXERC_year": "DT_FIM_EXERC"}, axis=1)
profit_year = profit_year.drop(["VL_CONTA_year", "VL_CONTA_quarters", "DT_FIM_EXERC_quarters"], axis=1)

profit_quarter = pd.concat([profit_quarter, profit_year]).sort_values(by="DT_INI_EXERC").reset_index()
```
