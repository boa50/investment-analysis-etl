```python
import pandas as pd
import os
```

```python
def load_files(years_load, files_load):
    df = pd.DataFrame()

    for year in years_load:
        for file in files_load:
            fname = "../data/raw/" + file + str(year) + ".csv"

            if os.path.isfile(fname):
                df_tmp = pd.read_csv(
                            fname,
                            encoding="ISO-8859-1",
                            sep=";",
                        )
                
                df_tmp.loc[:, "FILE_NAME"] = file
                
                df = pd.concat([df,df_tmp])
            else:
                print(fname + " not found! Skipping")

    return df
```

```python
years_load = [2024]
files_load = [
    "itr_cia_aberta_BPA_ind_",
    "itr_cia_aberta_BPP_ind_",
    "itr_cia_aberta_DFC_MD_ind_", 
    "itr_cia_aberta_DFC_MI_ind_", 
    "itr_cia_aberta_DMPL_ind_", 
    "itr_cia_aberta_DRA_ind_", 
    "itr_cia_aberta_DRE_ind_", 
    "itr_cia_aberta_DVA_ind_"
    ]
```

```python
df = load_files(years_load, files_load)
```

```python
df_fields = df.copy()
df_fields["DS_CONTA"] = df_fields["DS_CONTA"].str.lower()
df_fields = df_fields.groupby(["FILE_NAME","CD_CONTA", "DS_CONTA"]).count().reset_index()
df_fields = df_fields[["FILE_NAME", "CD_CONTA","DS_CONTA"]]
df_fields.head()
```
